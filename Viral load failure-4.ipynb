```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb

# Load dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models with custom names
model_names = ["LR", "SVM", "RF", "KNN", "DT", "XGB"]
models = {
    model_names[0]: LogisticRegression(max_iter=10000),
    model_names[1]: SVC(probability=True),
    model_names[2]: RandomForestClassifier(),
    model_names[3]: KNeighborsClassifier(),
    model_names[4]: DecisionTreeClassifier(),
    model_names[5]: xgb.XGBClassifier(eval_metric='logloss')
}

# Initialize lists to store metrics
results = []

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Compute Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    results.append({
        'Model': name,
        'Accuracy': accuracy
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Plotting
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Define a color palette for the models
colors = sns.color_palette("tab10", n_colors=len(results_df))

# Create bar plot for Accuracy
bar_plot = sns.barplot(x='Model', y='Accuracy', data=results_df, palette=colors)
plt.ylabel('Accuracy')
plt.title('Model Comparison: Accuracy')
plt.xticks(rotation=45)
plt.ylim(0, 1)  # Accuracy ranges from 0 to 1

# Add value annotations on bars
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.2f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 9), 
                      textcoords='offset points')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier  # Import KNeighborsClassifier
import xgboost as xgb

# Load dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models
models = {
    'XGBoost': xgb.XGBClassifier(eval_metric='logloss'),
    'SVM': SVC(probability=True),
    'Random Forest': RandomForestClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=10000),
    'Decision Tree': DecisionTreeClassifier(),
    'KNearest Neighbour': KNeighborsClassifier()  # Correctly use KNeighborsClassifier
}

# Initialize lists to store metrics
results = []

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    y_pred = model.predict(X_test)
    
    # Compute ROC AUC
    auc = roc_auc_score(y_test, y_pred_prob)
    
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    # Compute Sensitivity (Recall) and Specificity
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    
    results.append({
        'Model': name,
        'Sensitivity': sensitivity,
        'Specificity': specificity,
        'AUC': auc
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))
sns.set(style="whitegrid")

# Create bar plots for Sensitivity and Specificity
results_df.set_index('Model')[['Sensitivity', 'Specificity']].plot(kind='bar', ax=ax1, color=['blue', 'red'], alpha=0.7)
ax1.set_ylabel('Score')
ax1.set_title('Model Comparison: Sensitivity, Specificity, and AUC')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
ax1.legend(loc='upper left')

# Create a secondary y-axis for AUC
ax2 = ax1.twinx()
sns.lineplot(x='Model', y='AUC', data=results_df, marker='o', color='green', ax=ax2, label='AUC')
ax2.set_ylabel('AUC', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add legend for the AUC line plot
handles, labels = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles + handles2, labels + labels2, loc='upper left')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb

# Load dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models
models = {
    'XGBoost': xgb.XGBClassifier(eval_metric='logloss'),
    'SVM': SVC(probability=True),
    'Random Forest': RandomForestClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=10000),
    'Decision Tree': DecisionTreeClassifier(),
    'KNearest Neighbour': KNeighborsClassifier()
}

# Initialize lists to store metrics
results = []

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Compute Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    results.append({
        'Model': name,
        'Accuracy': accuracy
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Plotting
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Define a color palette for the models
colors = sns.color_palette("coolwarm", n_colors=len(results_df))

# Create bar plot for Accuracy
bar_plot = sns.barplot(x='Model', y='Accuracy', data=results_df, palette=colors)
plt.ylabel('Accuracy')
plt.title('Model Comparison: Accuracy')
plt.xticks(rotation=45)
plt.ylim(0, 1)  # Accuracy ranges from 0 to 1

# Add value annotations on bars
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.2f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', va = 'center', 
                      xytext = (0, 9), 
                      textcoords = 'offset points')

plt.show()
```
